## **2. Demonstrate data splitting for training and testing data. Using any preprocessing library in Python on the iris dataset provided in the module.**

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/drive/MyDrive/Assignment 1.1 ML Tools, Process, and Your Learnings/Iris_Data.csv', encoding = "latin-1")
df.head()
```

```python
# seperate the dataset into x and y
X = df.drop(columns = ['species'])
y = df.species
X
```

```python
y
```

#### Data Splitting

```python
#encodes y as numeric instead of string
from sklearn.preprocessing import LabelEncoder
y_encoded = LabelEncoder().fit_transform(y)
```

```python
#split the dataset (20% of test)
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, random_state=5, test_size=0.2)
```

```python
X_train.head()
```

```python
X_train.describe()
```

```python
X_test.head()
```

```python
X_test.describe()
```

#### Model

```python
# import a logistic regression for classification, linear regression outputs a decimal (harder to classify).
from sklearn.linear_model import LogisticRegression
iris_model = LogisticRegression(max_iter=200)
```

```python
iris_model.fit(X_train, y_train)
```

```python
X_test
```

```python
iris_model.predict(X_test)
```

```python
y_test
```

```python
iris_model.score(X_test, y_test)
```

97% accuracy of the model

## **3. Provide a sample confusion matrix and demonstrate the computation of TP, FP, FN, TN. How can we derive accuracy of the model based on the previously mentioned metrics?**

I used the iris dataset for the confusion matrix. The computation of TP, FP, FN, TN is done in the library confusion_matrix using the values of y and the predicted values of y

```python
from sklearn.metrics import confusion_matrix
y_pred = iris_model.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm_plot = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['setosa', 'versicolor', 'virginica'])
cm_plot.plot(cmap='autumn_r')
```



> True positive and True negatives denotes the correct classifications in the predictions. False Positive and False Negative are the incorrect classications based on their mistakes. Through this metrics, we can check the pattern of mistakes in our model, like if the model is often have false positives. Then, we can take action to solve the problem of the model through the pattern that we can see.






> Through the metrics of the Confusion Matrix, you can get the overall accuracy of your model by using this formula.

$$
\text{Accuracy} = \frac{TP + TN}{TP + FP + FN + TN}
$$
